```python
import pandas as pd
```

```python
df = pd.read_excel("新疆_别斯铁热克风电场_2017_风电场动态数据表.xls")
```

```python
windmid = df["测风塔10m风速(m/s)"].rolling(4).mean()
```

```python
windmax = df["测风塔10m风速(m/s)"].rolling(4).max()
```

```python
windmin = df["测风塔10m风速(m/s)"].rolling(4).min()
```

```python
t = [5+4*i for i in range(24)]
```

```python
len(t)
```

```python
windmid = windmid.loc[t]
```

```python
windmax = windmax.loc[t]
```

```python
windmin = windmin.loc[t]
```

```python
windmid.plot()
```

```python
%matplotlib inline
```

```python
new = pd.DataFrame()
```

```python
new["mid"] = windmid
```

```python
new["max"] = windmax
new["min"] = windmin
```

```python
new.plot()
```

```python
new.to_csv("1.csv")
```
